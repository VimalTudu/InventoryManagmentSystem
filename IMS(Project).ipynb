## Display function:

```python
#It is used to display product description present in database

def display():
    fout=open("records.json","r")
    data=fout.read()
    if(data==""):
        data+='{}'
    fout.close()
    d_data=json.loads(data)
    
    #When there is no product in database records.json
    if(len(d_data)==0):
        print("\nSorry No Product")
        
        
    #Printing details of database
    for d in d_data:
        print("\nProduct_id:",d,d_data[d])
        
```

## User Input Function: 

```python
#This function takes input from user
#It also return nested dictionary containing extra attribute 'user name' and 'user_quantity'->which is the entered by user(Customer)

def user_input():
    
    user_name=input("\nEnter customer name: ")
    item=input("Enter no. of items to purchase:")
    
    #file(records.json) is read and data is converted into dictionary js
    #records.json-->data(text)-->js(dict)
    
    fout=open("records.json","r")
    data=fout.read()
    if(data==""):
        data+='{}'
    fout.close()
    js=json.loads(data)
    
    #purchase data is nested dictionary 
    purchase_data={}
    purchase_data["User_Name"]=user_name #Adding extra key and value pairs

    #if user has entered item greater than the total items present in database 
    if(str(item)=="" or int(item)>len(js) or int(item)<=0  ):
        print("Please Enter valid no. of item")
        
    ##Incase user enter nothing    
    if(item==""): 
        item=0
    else:
        #Typecasting 
        item=int(item)
    
    for i in range(item):
        
        p_id=input("Enter product id: ")
        
        #if product id is not in database
        if(p_id not in js):
            print("Enter valid product id")
            return purchase_data
        
        qu=int(input("Enter the quantity: "))
        
        #Out of stock
        if(int(js[p_id]['quantity'])<=0):
            print("Product ID:{} Product Name:{} Out of stock".format(p_id,js[p_id]['name']))
            return purchase_data
        
        #if quantity entred by user is greater than the available in database
        if(int(qu)>int(js[p_id]['quantity'])):
            print("Please enter quantity less than previous")
            print("Bill will be created from above entered product id")
            return purchase_data
        
        if(int(qu)<=0):
            print("Enter valid quantity,It should be greater than 0")
            print("Bill will be created from above entered product id")
            return purchase_data
        
        purchase_data[p_id]=js[p_id]
        
        #Adding extra attribute key_>value pairs inside nested dictionary to particular product_id 
        purchase_data[p_id]['user_quantity']=qu         
        
        ##print(purchase_data)
    return(purchase_data)
    #return(list([user_name,{p_id:js[p_id],'user_quantity':qu}]))
```

```python
#This function is used to print the reciept:
#user_data is the purchase_data returned from user_input() function
def billing(user_data):
    
    #print(user_data)
    
    
    if(len(user_data)<2):
        print("==================================")
        print("No Purchase")
        return
    
    
    #To get unique refrence id
    #---------------------------------------------------------------------------------------------------------
    #Opening sales.json and converting it into dictionary
    #sales.json-->sal_txt(text)-->sales_data(dict)
    
    sales_js=open("sales.json","r")
    sal_txt=sales_js.read()
    if(sal_txt==""):
        sal_txt+='{}'
    sales_js.close()
    
    #Converting sales data(sales.json) from text to dictionary
    sales_data=json.loads(sal_txt)
    
    #uid-->Unique referenceid
    
    #Adding new unique refrence id of purchase
    uid=len(sales_data)
    #---------------------------------------------------------------------------------------------------------
    
    
    #Print bill details
    
    print("\n************THE SHOP*********************")
    print("*************Ranchi,Jharkhand************")
    print("\nBill Details\n")
    print("========================================\n")
    print("Customer name:",user_data["User_Name"])
    print("Refrence id:".format(uid))
    #print(user_data.keys())
    
    for u in user_data:
        if(u=='User_Name'):
            continue
        print("\nProduct ID:",u,end="  ")
        print("Product Name:",user_data[u]['name'],end="  ")
        print("Price:\u20B9{}".format(user_data[u]['mrp']),end="  ")
        print("Quantity:",user_data[u]['user_quantity'],end=" ")
        print("Cost:\u20B9{}".format(float(user_data[u]['mrp']*user_data[u]['user_quantity'])))
    
    print("------------------------------------------------")
    print("Date and Time:",time.ctime())
    #print("Product details:\n")
    #print(user_data[3:])
    print("\n----------------------------------------------")
    
    
    
    cost=0.0
    gst=0.0
    total_cost=0.0
    for u in user_data:
        if(u=='User_Name'):
            continue
        cost+=user_data[u]['mrp']*user_data[u]['user_quantity']
    #print(user_data[3]['mrp']
    
    gst=(18*cost)/100
    total_cost=cost+gst
    print("-------------------------------------------------")
    print("Total items:",len(user_data)-1)
    print("Cost:\u20B9{}".format(float(cost)))
    print("GST:\u20B9{}".format(float(gst)))
    print("Total Amount(including all tax):\u20B9{}".format(float(total_cost)))
    print("-------------------------------------------------")
    print("THANK YOU! VISIT AGAIN")
    
    #print(cost,gst,total_cost)
    #process()
```

## Process function:

```python
#Function to process data which updates the database

def process(user_data):
    
    #file is opend and data is converted to dictionary for processing
    #records.json.json-->data(text)-->update_data(dict)
    
    fin=open("records.json","r")
    data=fin.read()
    if(data==""):
        data+='{}'
    fin.close()
    update_data=json.loads(data)
    
    #sf_text is the text which is written in sales.txt
    
    ##sf=open("sales.txt","w")
    sf_txt="\n\nCustomer Name: "+user_data['User_Name']
    
    #Opening sales.json and converting it into dictionary
    #sales.json-->sal_txt(text)-->sales_data(dict)
    
    sales_js=open("sales.json","r")
    sal_txt=sales_js.read()
    if(sal_txt==""):
        sal_txt+='{}'
    sales_js.close()
    
    #Converting sales data(sales.json) from text to dictionary
    sales_data=json.loads(sal_txt)
    
    
    #uid-->Unique referenceid
    
    #Adding new unique refrence id of purchase
    uid=len(sales_data)
    #print(sal_txt)
    #print("Uid",uid)
    #sales_data[uid]=len(sales_data)
    sales_data[uid]=dict()
    sales_data[uid]['Customer_Name']=user_data['User_Name']
    sales_data[uid]['Date_Time']=time.ctime()
    
    
    sf_txt+=" Refrence id:{} DateTime:{}".format(uid,time.ctime())
    
    #cost-->sub_total(cost) of each product ,total_cost-->total cost of all products purchased
    cost=0.0
    total_cost=0.0
    
    for u in user_data:
        if(u=='User_Name'):
            continue
            
        #Changing values of 'quantity' of product
        update_data[u]['quantity']=user_data[u]['quantity']-user_data[u]['user_quantity']
        
        
        
        #To verify the changes if necessary
        ''' #print("#")
            #print("Original Quantity:",user_data[u]['quantity'])
        #print("User_Quantity:",user_data[u]['user_quantity'])
        #print("Updated Quantity:", update_data[u]['quantity'])
        #print("#")'''
        
        total_cost=cost+user_data[u]['user_quantity']*user_data[u]['mrp']
        cost=user_data[u]['user_quantity']*user_data[u]['mrp']
        
        sf_txt+='''\nProduct_id:{} Quantity:{} User_Quantity:{} Updated Quantity:{} Cost:{}'''.format(u,user_data[u]['quantity'],user_data[u]['user_quantity'],update_data[u]['quantity'],cost)
        
        sales_data[uid][u]=dict()
        sales_data[uid][u]['Product_id']=u
        sales_data[uid][u]['Quantity']=user_data[u]['user_quantity']
        sales_data[uid][u]['Cost']=cost
        
    #File(records.json.json) is opened and necessary update is done(CHANGE IN ORIGINAL DATABASE-->records.json.json)
    fout=open("records.json.json","w") 
    update_data=json.dumps(update_data,indent=2)
    fout.write(update_data)#Updating the quantity
    fout.close()
    
    #file sales.txt is updated(Changes in sales.txt)
    sf_txt+="\nEarning:{}".format(float(total_cost))
    sf=open("sales.txt","a")     
    sf.write(sf_txt)
    sf.close()
    
    sales_data[uid]['Total_Amount']=total_cost
    
    
    #file sales.json is updated(Changes in sales.json)
    sales_js=open("C:/Users/ASUS/Desktop/sales.json","w")
    sales_js.write(json.dumps(sales_data,indent=2))
    sales_js.close()
    
    
    
```

## Purchase Function:

```python
##It calls 4 function: display(),user_input(),billing(),process()
## user_data --> stores the dictionary returned from user_input() which has details about cutomer and product purchased by him/her.


def purchase():
    display()
    print("-----------------------------------------------------------------------------")
    print("Please Choose from the above products\n")
    print("-----------------------------------------------------------------------------")
    user_data=user_input()
    billing(user_data)
    process(user_data)
```
